```
import requests
from bs4 import BeautifulSoup
import json
import os

q = "風景"
page = 1
number = 100
h = {
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36",    
}
imglist = []
while page < 6:
    url = "https://www.google.com/search?ei=5tGJXbnxFu-xmAXko6K4Ag&yv=3&tbm=isch&q={}&vet=10ahUKEwi5_MW4g-nkAhXvGKYKHeSRCCcQuT0IjwIoAQ.5tGJXbnxFu-xmAXko6K4Ag.i&ved=0ahUKEwi5_MW4g-nkAhXvGKYKHeSRCCcQuT0IjwIoAQ&ijn={}&start={}&asearch=ichunk&async=_id:rg_s,_pms:s,_jsfs:Ffpdje,_fmt:pc".format(q, page, number)
    response = requests.get(url, headers= h)
    html = BeautifulSoup(response.text)

    divs = html.find_all("div", class_="rg_meta")

    if len(divs)==0:
        print("目前頁面可能會最後一頁")
        break

    for i in divs:
        imglist.append(json.loads(i.text)["ou"])
    page += 1
    number += 100

filename = "./views/"
if not os.path.exists(filename):
    os.makedirs(filename)

# 搜尋檔案
basefile = ["jpg", "jpeg", "png"]
for i, j in enumerate(imglist):
    for k in basefile:
        if k.upper() in j or k in j:
            try:
                # 圖片抓取方法
                img = requests.get(j, stream=True, verify=False)
                with open(filename+"{}.{}".format(i, k), "wb") as f:
                    f.write(img.raw.read())
            except:
                print("放棄:{}".format(j))


```
