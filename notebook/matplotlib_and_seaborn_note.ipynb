```python
'''

先給總圖，在畫子圖

'''

fig = plt.figure()
# 可調整子圖之間彼此的間距
fig.subplots_adjust(hspace=1, wspace=1)
nrow = 2
ncol = 3
for order in range(6):
    plt.subplot(nrow, ncol, order + 1)
    plt.text(0.5,0.5,f"{nrow, ncol, order + 1}",fontsize=12,ha="center")
plt.show()
```

```python
'''

直接畫子圖


'''

nrow = 2
ncol = 3

# 注意此處sharex="col"及sharey="row"不可以把"col"跟"row"交換喔，不然會沒效果~~
fig, ax = plt.subplots(nrow, ncol, sharex="col", sharey="row")

print("=====我是分隔線")
print(type(ax))
print(ax)
print("=====我是分隔線")

# 注意此處沒有給大畫布(plt.figure)，故沒有subplot_adjust此性質。
# fig.subplot_adjust(hspace=0.4, wspace=0.4)

for row in range(2):
    for col in range(3):
        ax[row, col].text(0.5, 0.5, f"{row, col}", ha="center")
        ax[row, col].set_title(f"{row, col}")
        
plt.show()
```

```python
# example
mean = [0,0]
cov = [[1,1],[1,2]]

rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, 100).T

nrow = 4
ncol = 4
fig = plt.figure(figsize=(8,8))
grid = plt.GridSpec(nrow, ncol, hspace=0.4, wspace=0.4)

main_ax = fig.add_subplot(grid[:-1,1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[],sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], sharex=main_ax)

main_ax.plot(x, y, "ok", markersize=5)
y_hist.hist(y, bins=40, histtype="stepfilled", orientation="horizontal", color="gray")
y_hist.invert_xaxis()
x_hist.hist(x, bins=40, histtype="stepfilled", color="gray")
x_hist.invert_yaxis()
```

```python
import pandas as pd
import seaborn as sns

dict_ = {
    "country":["Taiwan","China","China","China","Taiwan"],
    "continent":["Asia","Asia","Asia","Asia","Asia"],
    "Year":[1992,1960,1988,1990,1943],
    "Life_Exp":[28.80,30.33,31.99,34.02,36.08],
    "pop":[8400500,9876543,10765431,11234567,12987650],
    "GDP_per_CAP":[779.44,820.53,853.10,836.19,739.98]
}

df = pd.DataFrame(dict_)
df["popLog"] = np.log10(df["pop"])
df["GDP_per_CAP_Log"] = np.log10(df["GDP_per_CAP"])
df = df.drop(["pop", "GDP_per_CAP"], axis=1)
df
```

```python
'''

定義等等畫圖會使用到的函式

'''

def Corr(x, y, **kwargs):
    corr = np.corrcoef(x, y)[0, 1]
    
    mark = r"$\rho$ = {:.2f}".format(corr)
    
    # 取得當前axes
    ax = plt.gca()
    
    # 為啥這段要加 xycoords才會顯示出圖來?
    ax.annotate(mark, xy=(0.3, 1), fontsize=20, xycoords = ax.transAxes)

def Describe(col, **kwargs):
    
    # Statistics
    mean = np.mean(col)
    max = np.max(col)
    min = np.min(col)
    Q50 = np.quantile(col, q=0.5)
    std = np.std(col)
    
    # text
    mark = f'''
        mean : {mean:.2f}
        std  : {std:.2f}
        max  : {max:.2f}
        min  : {min:.2f}
        Q50  : {Q50:.2f}
    '''   
    
    # 取得當前axes
    ax = plt.gca()
    ax.annotate(mark, xy=(0.1, 0.1), fontsize=12, xycoords = ax.transAxes)

'''

開始畫圖...

'''
plt.figure(figsize=(10,10))

grid = sns.PairGrid(df, vars=["Life_Exp", "popLog", "GDP_per_CAP_Log"])
# 也可直接統一使用grid.map() 統一畫圖
grid.map_upper(plt.scatter, color="red")
grid.map_upper(Corr)
grid.map_lower(sns.kdeplot, cmap="YlGnBu")
grid.map_diag(Describe) # 怎取得欄位名稱???

# 要有figure才能使用suptitle
plt.suptitle("\n".join(["Look Scatter and hist"] * 2 ), fontsize=28, y=1.15)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# 不懂為什麼要新增這個欄位???
tips["tip_pct"] = tips["tip"] * 100 / tips["total_bill"]

with sns.axes_style("white"):
    grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
    grid.map(plt.hist, "tip_pct", bins=np.linspace(0,100,50))

with sns.axes_style("ticks"):
    ax = sns.factorplot("day", "tip_pct", "sex", data=tips, kind="box")
    ax.set_axis_labels("day", "tip_pct")
    
with sns.axes_style("white"):
    sns.jointplot("total_bill", "tip", data=tips, kind="hex")
    
```

```python
ax = sns.jointplot("total_bill", "tip", data=tips, kind="reg")
ax.ax_joint.plot(np.linspace(0,60,1000), np.linspace(0,10,1000), linestyle=":", color="k")
```

```python
ax = sns.lmplot("total_bill", "tip", col="sex",data=tips, markers=".", scatter_kws={"color":"c"})
ax.map(plt.axhline, y=2, color="k", linestyle=":")
```
